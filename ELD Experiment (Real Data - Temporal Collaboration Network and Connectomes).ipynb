```python
## Import all relevant libraries
import sys
import networkx as nx 
import numpy as np
import seaborn as sns
from ELD import create_ELD_distance_matrix
```

```python
# Load temporal IBM collaboration networks data
# The data used here is downloaded from this link https://doi.org/10.6084/m9.figshare.7893002
# which is provided by the authors of this paper:
# https://appliednetsci.springeropen.com/articles/10.1007/s41109-019-0156-x#Sec10

G2013 = nx.read_graphml("data/dev-dev-weighted_full__year2013.graphml")
G2014 = nx.read_graphml("data/dev-dev-weighted_full__year2014.graphml")
G2015 = nx.read_graphml("data/dev-dev-weighted_full__year2015.graphml")
G2016 = nx.read_graphml("data/dev-dev-weighted_full__year2016.graphml")
G2017 = nx.read_graphml("data/dev-dev-weighted_full__year2017.graphml")
```

```python
K = 20
graphList = [G2013, G2014, G2015, G2016, G2017]
distance_mat = create_ELD_distance_matrix(graphList, K)
names = ["2013", "2014", "2015", "2016", "2017"]
ax = sns.heatmap(distance_mat, xticklabels = names, yticklabels = names, annot = True, fmt=".2f")
ax.set_title("Distance Matrix: Citation Graphs from 2013 - 2017, k = 20")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_yticklabels(ax.get_yticklabels(), rotation=45)
fig = ax.get_figure()
fig.savefig("Real_data_2013-2017", bbox_inches = "tight")
```

```python
## The connectomes data used here is downloaded from this neurodata link: 
## https://neurodata.io/project/connectomes/

pacificus1 = nx.read_graphml("data/p.pacificus_neural.synaptic_1.graphml")
pacificus2 = nx.read_graphml("data/p.pacificus_neural.synaptic_2.graphml")
rat1 = nx.read_graphml("data/rattus.norvegicus_brain_1.graphml")
rat2 = nx.read_graphml("data/rattus.norvegicus_brain_2.graphml")
rat3 = nx.read_graphml("data/rattus.norvegicus_brain_3.graphml")
mouse1 = nx.read_graphml("data/mouse_visual.cortex_1.graphml")
mouse2 = nx.read_graphml("data/mouse_visual.cortex_2.graphml")
```

```python
K = 20
graphList = [pacificus1, pacificus2, rat1, rat2, rat3, mouse1, mouse2]
graphList = [graph.to_undirected() for graph in graphList]
distance_mat = create_ELD_distance_matrix(graphList, K)
names = ["pacificus1", "pacificus2", "rat1", "rat2", "rat3", "mouse1", "mouse2"]
ax = sns.heatmap(distance_mat, xticklabels = names, yticklabels = names, annot = True, fmt=".2f")
ax.set_title("Distance Matrix: Connectomes of different organisms, k = 20")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_yticklabels(ax.get_yticklabels(), rotation=45)
fig = ax.get_figure()
fig.savefig("Real_data_connectomes", bbox_inches = "tight")
```
