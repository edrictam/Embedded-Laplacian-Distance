```python
## Import all relevant libraries
import sys
import networkx as nx 
import numpy as np
import seaborn as sns
from ELD import create_ELD_distance_matrix
```

```python
import random

random.seed(10)
np.random.seed(10)

## Define helper function to generate random weighted ER graphs (edge weights exponentially distributed)
def getERWeighted(n, p, scale):
    G= nx.erdos_renyi_graph(n, p, seed=None, directed=False)
    for (u,v,w) in G.edges(data=True):
        w['weight'] = np.random.exponential(scale)
    return G
```

```python
## Create different families of graphs with different parameters

graphlist1 = [getERWeighted(20, 0.5, 30) for i in range(2)]
graphlist2 = [getERWeighted(20, 0.5, 2) for i in range(2)]
graphlist3 = [getERWeighted(100, 0.5, 30) for i in range(2)]
graphlist4 = [getERWeighted(100, 0.5, 2) for i in range(2)]

graphlist5 = [getERWeighted(20, 0.8, 20) for i in range(2)]
graphlist6= [getERWeighted(20, 0.2, 20) for i in range(2)]
graphlist7 = [getERWeighted(100, 0.8, 20) for i in range(2)]
graphlist8 = [getERWeighted(100, 0.2, 20) for i in range(2)]


```

```python
K = 5
graphList = graphlist1 + graphlist2 + graphlist3 + graphlist4  
distance_mat = create_ELD_distance_matrix(graphList, K)
names = ["ER20,S30","ER20,S30", "ER20,S2","ER20,S2", "ER100,S30", "ER100,S30", "ER20,S2","ER20,S2"]
sns.set(font_scale=1)
ax = sns.heatmap(distance_mat, xticklabels = names, yticklabels = names, annot = True, fmt=".1f")
ax.set_title("Distance Matrix: ER graphs w/ Exp. Dist. Weights, k = 5")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_yticklabels(ax.get_yticklabels(), rotation=45)
fig = ax.get_figure()
fig.savefig("ER_scale_k_5.png", bbox_inches = "tight")
```
