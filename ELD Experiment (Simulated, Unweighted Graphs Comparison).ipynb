```python
K = 5
graphList = [P10, P20, P30, P40, P50, P100, P200, P500]
distance_mat = create_ELD_distance_matrix(graphList, K)
names = ["P10", "P20", "P30", "P40", "P50", "P100", "P200", "P500"]
ax = sns.heatmap(distance_mat, xticklabels = names, yticklabels = names, annot = True, fmt=".3f")
ax.set_title("Distance Matrix Heatmap for Path Graphs, k = 5")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_yticklabels(ax.get_yticklabels(), rotation=45)
fig = ax.get_figure()
fig.savefig("Path_k_5.png")
```

## Below, we use the NPD distance to compare the same Rings of Cliques family as above 
## The resulting NPD distance matrix heatmap can be compared to the ELD distance matrix heatmap above

```python
from portrait_divergence import portrait_divergence

def create_dist_matrix_NPD(graphList):
    dist_matrix = np.zeros([len(graphList), len(graphList)])
    for i, G1 in enumerate(graphList):
        for j, G2 in enumerate(graphList):
            dist_matrix[i,j] = portrait_divergence(G1, G2)
    dist_final = dist_matrix
    return dist_final
```

```python
graphList = [R33, R35, R53,  R55, R63, R65, R103,  R105]
distance_mat = create_dist_matrix_NPD(graphList)
names = ["R3,3", "R3,5", "R5,3",  "R5,5", "R6,3", "R6,5" ,"R10,3", "R10,5"]
ax = sns.heatmap(distance_mat, xticklabels = names, yticklabels = names, annot = True, fmt=".3f")
ax.set_title("NPD Distance Matrix Heatmap for Ring of Cliques Graphs")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_yticklabels(ax.get_yticklabels(), rotation=45)
fig = ax.get_figure()
fig.savefig("NPD_RingCliques.png")
```
